**Example 1**: Write a program to enter a number and display its hex and octal equivalent and its square root.

```python
number = int(input("Enter a number: "))
print("HEX of " + str(number) + " is " + str(hex(number)))
print("OCT of " + str(number) + " is " + str(oct(number)))
print("Square root of " + str(number) + " is " + str(number**0.5))
```

**Example 3**: Write a program to calculate area of a triangle using Heron's formula.

```python
# Hint - Heron's formula is area = sqrt(s*(s-a)*(s-b)*(s-c)), where s=(a+b+c)/2
a = float(input("enter the value for a"))
b = float(input("enter the value for b"))
c = float(input("enter the value for c"))
s = (a+b+c)/2
area = (s*(s-a)*(s-b)*(s-c))**0.5
print(str(area))
```

**Example 4**: Write a program to calculate the distance between two points.

```python
x1 = float(input("enter the value for x1"))
x2 = float(input("enter the value for x2"))
y1 = float(input("enter the value for y1"))
y2 = float(input("enter the value for y2"))
distance = ((x2-x1)**2 + (y2-y1)**2)**0.5
print(distance)
```

**Example 5**: Write a program to perform addition, subtraction, multiplication, division, integer division, and modulo division on two integer numbers.

```python
a = int(input("enter the first integer")) #100
b = int(input("enter the second integer")) #10

add = a + b
sub = a - b
multi = a * b
idiv = a / b
modiv = a % b
fdiv = float(a) / b

print(str(a) + "+" + (str(b)) + "=" + str(add))
print(str(a) + "-" + (str(b)) + "=" + str(sub))
print(str(a) + "*" + (str(b)) + "=" + str(multi))
print(str(a) + "/" + (str(b)) + "=" + str(idiv))
print(str(a) + "%" + (str(b)) + "=" + str(modiv))
print(str(a) + "/" + (str(b)) + "=" + str(fdiv))
```

**Example 13**: Write a program to calculate the bill amount for an item given its quantity sold, value, discount and tax.

```python
qty = float(input("quantity"))
val = float(input("value"))
disc = float(input("discount"))
tax = float(input("tax"))

amount = qty*val
discountedAmount = amount*disc/100
subtotal = amount-discountedAmount
taxAmount = subtotal*tax/100
finalprice = subtotal+taxAmount

print(finalprice)
```
